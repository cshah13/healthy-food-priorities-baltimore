# Import Libraries


```python
# import libraries

# data analysis
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import plotly.express as px

# download files to our computer
from google.colab import files 
```

# Import Data for Healthy Food Quality Index

Imported data from [this github here](https://github.com/cshah13/food-index-baltimore-change)




```python
# import csv of healthy food quality index data

# save github csv link
index_data = "https://raw.githubusercontent.com/cshah13/food-index-baltimore-change/main/HFAI%20Python%20Data.csv"

#define our initial dataframe
df_index = pd.read_csv(index_data)
```

# Look at out data

```python
# preview the first five rows
df_index.head()
```

```python
df_index = df_index.rename(columns = {"CSA2010" : "Location", "hfai12" : "HFAI 2012", "hfai15" : "HFAI 2015"})
```

```python
# plot indexes for 2012 and 2015 in a double line graph
plot = df_index.plot(x = "Location", y = ['HFAI 2012', 'HFAI 2015'], kind = "line", title = "Healthy Food Index Scores in Baltimore, MD in 2012 and 2015", figsize = (18,8))

plt.xlabel("Areas in Baltimore")
plt.ylabel("Healthy Food Availability Index Score")


```

```python
# plot indexes for 2012 and 2015 in a double line graph
plot = df_index.plot(x = "Location", y = ['HFAI 2012', 'HFAI 2015'], kind = "bar", title = "Healthy Food Index Scores in Baltimore, MD in 2012 and 2015", figsize = (15,8))

plt.xlabel("Areas in Baltimore")
plt.ylabel("Healthy Food Availability Index Score")
```

```python
# make double bar graph
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import plotly.graph_objects as px
import numpy as np

np.random.seed(56)  


random_x= np.random.randint(1,101,100) 
random_y= np.random.randint(1,101,100)
  
x = df_index['Location']
  
plot = px.Figure(data=[go.Bar(name = 'HFAI 2012', 
                              x = df_index['Location'],
                              y = df_index['HFAI 2012']),
                       go.Bar(name = 'HFAI 2015', x = df_index['Location'],y = df_index['HFAI 2015'])])

plot.update_layout(title="Health Food Availability Index in Baltimore, MD: 2012 and 2015", xaxis_title="Locations",
                   yaxis_title="Healthy Food Availability Index", width=1000, height=600)

plot.update_xaxes(categoryorder="total ascending")
plot.show()
```
