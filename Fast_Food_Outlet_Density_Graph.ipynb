## Import Libraries 


```python
# import libraries
import pandas as pd

# data analysis
# data visualizations
import matplotlib.pyplot as plt
import plotly.express as px

# download files to our computer 
from google.colab import files 
```

```python
#import csv of Fast Food Data
# define our initial dataframe
df_food = pd.read_csv("https://raw.githubusercontent.com/apate139/Final-Project-Colab/main/Fast_Food_Outlet_Density_per_1%252C000_Residents.csv")
df_food.head()
df_food = df_food.rename(columns = {"CSA2010" : "Location", "Fastfd13" : "Fast Food Outlet Density Per 1000 Residents 2013", "Fastfd18" : "Fast Food Outlet Density Per 1000 Residents 2018"})
```

## Visualize Data 

```python


plot = df_food.plot(x = "Location", y = ['Fast Food Outlet Density Per 1000 Residents 2013', 'Fast Food Outlet Density Per 1000 Residents 2018'], kind = "bar", title = "Fast Food Outlet Density Per 1000 Residents by Community Statistical Area in Baltimore, MD, 2013 and 2018", figsize = (15,5))

plt.xlabel("Community Statistical Areas")
plt.ylabel("Fast Food Outlet Density Per 1000 Residents")


```
